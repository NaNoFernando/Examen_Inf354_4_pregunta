```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
#para graficar el arbol de desicion
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
data=pd.read_csv("datasetVideoGames2.csv")
data.columns
```

```python
variables_predictoras=data[['console', 'alcohol_reference', 'animated_blood', 'blood',
       'blood_and_gore', 'cartoon_violence', 'crude_humor', 'drug_reference',
       'fantasy_violence', 'intense_violence', 'language', 'lyrics',
       'mature_humor', 'mild_blood', 'mild_cartoon_violence',
       'mild_fantasy_violence', 'mild_language', 'mild_lyrics',
       'mild_suggestive_themes', 'mild_violence', 'no_descriptors', 'nudity',
       'partial_nudity', 'sexual_content', 'sexual_themes',
       'simulated_gambling', 'strong_janguage', 'strong_sexual_content',
       'suggestive_themes', 'use_of_alcohol', 'use_of_drugs_and_alcohol',
       'violence']]
variable_a_predecir=data.esrb_rating
variable_a_predecir.unique()#esta linea de codigo la usamos para obtener las salidas unicas del dataset
```

```python
#etiquetamos las predictoras y los objetivos
etiquetas_predictoras=['console', 'alcohol_reference', 'animated_blood', 'blood',
       'blood_and_gore', 'cartoon_violence', 'crude_humor', 'drug_reference',
       'fantasy_violence', 'intense_violence', 'language', 'lyrics',
       'mature_humor', 'mild_blood', 'mild_cartoon_violence',
       'mild_fantasy_violence', 'mild_language', 'mild_lyrics',
       'mild_suggestive_themes', 'mild_violence', 'no_descriptors', 'nudity',
       'partial_nudity', 'sexual_content', 'sexual_themes',
       'simulated_gambling', 'strong_janguage', 'strong_sexual_content',
       'suggestive_themes', 'use_of_alcohol', 'use_of_drugs_and_alcohol',
       'violence']
etiquetas_objetivos=["E","ET","M","T"]
```

```python
a=[]
b=[]
c=[]
d=[]
for i in range(11):
    x_entrenamiento,x_prueva,y_entrenamiento,y_prueva=train_test_split(variables_predictoras,variable_a_predecir,test_size=0.2)
    #aqui entrenamos al arbol
    arbol=DecisionTreeClassifier(criterion="entropy")
    arbol.fit(x_entrenamiento,y_entrenamiento)
    prediccion=arbol.predict(x_prueva)
    #verifica que tan segura es la prediccion del arbol, mientras mas cerca a 1 mucho mejor sera el clasificador que estamos usando
    a.append(arbol.score(x_prueva,y_prueva))
    b.append(arbol.score(x_entrenamiento,y_entrenamiento))
    c.append(1-arbol.score(x_prueva,y_prueva))
    d.append(1-arbol.score(x_entrenamiento,y_entrenamiento))
#muestra los indices de predicion del arbol
print("confiabilidad de los datos de prueva===========>  {} %".format(sum(a)/10))
print("error de los datos de prueva===================>  {} %".format(sum(c)/10))
print("confiabilidad de los datos de entrenamiento====>  {} %".format(sum(b)/10))#es mas que 1, entonces hay que hacer sobreajuste
print("error de los datos de entrenamiento============>  {} %".format(sum(d)/10))
```

```python
print(confusion_matrix(y_prueva,prediccion))
```
